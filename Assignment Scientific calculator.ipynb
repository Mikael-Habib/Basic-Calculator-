```python
import math
# fucntions for Addition

def Addition(y,z):
    Add = y+z
    return Add

def Subtraction(y,z):
    subtract = y-z
    return subtract

def Multiplication(y,z):
    Multiply = y*z
    return Multiply

def Division(y,z):
    divide = y/z
    return divide

def Power(y,z):
    Power_Num = pow(y,z)
    return Power_Num
def sqaure_root(y):
     SqRoNum = math.sqrt(y)
     return SqRoNum
def sine_func(y):
    y = y*(math.pi/180)
    Sine = math.sin(y)
    return Sine
def Cosine_func(y):
    y = y*(math.pi/180)
    Cosine = math.cos(y)
    return Cosine
def Tan_func(y):
    y = y*(math.pi/180)
    Tan = math.tan(y)
    return Tan
def sineh_func(y):
    #y = y*(math.pi/180)
    Sineh = math.sinh(y)
    return Sineh
def Cosineh_func(y):
    #y = y*(math.pi/180)
    Cosineh = math.cosh(y)
    return Cosineh
def Tanh_func(y):
    #y = y*(math.pi/180)
    Tanh = math.tanh(y)
    return Tanh
```

```python
print("Enter a number")
User_Input_1 = float(input())
print(User_Input_1)

while User_Input_1 != 0:
    print("Enter an Operation, {+ plus, - minus , * Multiply , / division,^ power,~ sqaure root , S sine function , C cosine function , T tan function, SH sineh function , CH cosineh function , TH tanh function , = to break  }")
    User_Input_2 = input()
    print(User_Input_2)
    if User_Input_2 =='+':
        print("Enter a number")
        User_Input_3 = float(input())
        print(User_Input_3)
        User_Input_1 = Addition(User_Input_1,User_Input_3)
        print('=',User_Input_1)
    elif User_Input_2 =='-':
        print("Enter a number")
        User_Input_3 = float(input())
        User_Input_1 = Subtraction(User_Input_1,User_Input_3)
        print('=',User_Input_1)
    elif User_Input_2 =='*':
        print("Enter a number")
        User_Input_3 = float(input())
        User_Input_1 = Multiplication(User_Input_1,User_Input_3)
        print('=',User_Input_1)
    elif User_Input_2 =='/':
        print("Enter a number")
        User_Input_3 = float(input())
        User_Input_1 = Division(User_Input_1,User_Input_3)
        print('=',User_Input_1)
    elif User_Input_2 == '^':
        print("Enter a number")
        User_Input_3 = float(input())
        User_Input_1 = Power(User_Input_1,User_Input_3)
        print('=',User_Input_1)
    elif User_Input_2 == '~':
        User_Input_1 = sqaure_root(User_Input_1)
        print('=',User_Input_1)

    elif User_Input_2 == 'S':
        User_Input_1 = sine_func(User_Input_1)
        print('=',User_Input_1)  
    elif User_Input_2 == 'C':
        User_Input_1 = Cosine_func(User_Input_1)
        print('=',User_Input_1)
    elif User_Input_2 == 'T':
        User_Input_1 = Tan_func(User_Input_1)
        print('=',User_Input_1)
    elif User_Input_2 == 'SH':
        User_Input_1 = sineh_func(User_Input_1)
        print('=',User_Input_1)  
    elif User_Input_2 == 'CH':
        User_Input_1 = Cosineh_func(User_Input_1)
        print('=',User_Input_1)
    elif User_Input_2 == 'TH':
        User_Input_1 = Tanh_func(User_Input_1)
        print('=',User_Input_1)

    elif User_Input_2 == '=':
        print(User_Input_1)
        break



    
```
